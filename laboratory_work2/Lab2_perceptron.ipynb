## Описание эксперимента

В этом эксперименте мы создадим модель многослойного персептрона (MLP)[Multilayer Perceptron](https://en.wikipedia.org/wiki/Multilayer_perceptron) используя  [Tensorflow](https://www.tensorflow.org/) для распознавания рукописных цифр.

Многослойный персептрон (MLP) представляет собой класс искусственных нейронных сетей с прямой связью. MLP состоит как минимум из трех слоев узлов: входного слоя, скрытого слоя и выходного слоя. За исключением входных узлов, каждый узел представляет собой нейрон, использующий нелинейную функцию активации. MLP использует метод обучения с учителем, называемый обратным распространением для обучения. Его многослойность и нелинейная активация отличают MLP от линейного персептрона. Он может различать данные, которые не являются линейно разделимыми.

Прежде чем мы просто попытаемся перейти от диапазона значений `[0...255]` к `[0...1]`.

## Построение модели

Мы будет строить последовательную [Sequential](https://www.tensorflow.org/api_docs/python/tf/keras/Sequential?version=stable) Keras модель с 4 слоями:

- Слой 1: [Flatten](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Flatten?version=stable) слой, который будет сглаживать (flatten) 2D матрицу изображений в одномерный вектор (1D vector).
- Слой 2: **Input(входной)** [Dense](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Dense?version=stable) слой состоит из 128 нейронов и функцией активации[ReLU](https://www.tensorflow.org/api_docs/python/tf/keras/activations/relu?version=stable).
- Слой 3: **Hidden(скрытый)** [Dense](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Dense?version=stable) слой состоит из 128 нейронов и функцией активации [ReLU](https://www.tensorflow.org/api_docs/python/tf/keras/activations/relu?version=stable).
- Слой 4: **Output(Выходной)** [Dense](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Dense?version=stable) слой состоит из 10 нейронов для вывода результата [Softmax](https://www.tensorflow.org/api_docs/python/tf/keras/activations/softmax?version=stable). Выходные данные представляют собой предположение сети. 0-й выход представляет вероятность того, что входная цифра равна «0», 1-й выход представляет вероятность того, что входная цифра равна «1» и так далее...

В этом примере мы использовали `kernel_regularizer` параметр слоя для управления переобучением модели. Другой распространенный подход к борьбе с переобучением может заключаться в использовании[dropout layers](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Dropout) (i.e. `tf.keras.layers.Dropout(0.2)`).

```python
model = tf.keras.models.Sequential()

# Input layers.
model.add(tf.keras.layers.Flatten(input_shape=x_train_normalized.shape[1:]))
model.add(tf.keras.layers.Dense(
    units=128,
    activation=tf.keras.activations.relu,
    kernel_regularizer=tf.keras.regularizers.l2(0.002)
))

# Hidden layers.
model.add(tf.keras.layers.Dense(
    units=128,
    activation=tf.keras.activations.relu,
    kernel_regularizer=tf.keras.regularizers.l2(0.002)
))

# Output layers.
model.add(tf.keras.layers.Dense(
    units=10,
    activation=tf.keras.activations.softmax
))
```

Здесь можно увидеть описание параметров построенной модели.

```python
model.summary()
```

Для построения схемы модели инсталируем `graphviz`.

```python
tf.keras.utils.plot_model(
    model,
    show_shapes=True,
    show_layer_names=True,
)
```

## Компиляция молели

```python
adam_optimizer = tf.keras.optimizers.legacy.Adam(learning_rate=0.001)

model.compile(
    optimizer=adam_optimizer,
    loss=tf.keras.losses.sparse_categorical_crossentropy,
    metrics=['accuracy']
)
```

## Обучение модели

```python
log_dir=".logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)

training_history = model.fit(
    x_train_normalized,
    y_train,
    epochs=5,
    validation_data=(x_test_normalized, y_test),
    callbacks=[tensorboard_callback]
)
```

Посмотрим, как менялась функция потерь в процессе обучения. Мы ожидаем, что в каждую следующую эпоху значение будет становиться все меньше и меньше.

```python
plt.xlabel('Epoch Number')
plt.ylabel('Loss')
plt.plot(training_history.history['loss'], label='training set')
plt.plot(training_history.history['val_loss'], label='test set')
plt.legend()
```

```python
plt.xlabel('Epoch Number')
plt.ylabel('Accuracy')
plt.plot(training_history.history['accuracy'], label='training set')
plt.plot(training_history.history['val_accuracy'], label='test set')
plt.legend()
```

## Использование модели для предсказания

Чтобы использовать модель, которую мы только что обучили для распознавания цифр, нам нужно вызвать метод `predict()` .

```python
predictions_one_hot = loaded_model.predict([x_test_normalized])
```

Каждый прогноз состоит из 10 вероятностей (по одной на каждое число от 0 до 9). Нам нужно выбрать цифру с наибольшей вероятностью, так как это будет цифра, в которой наша модель наиболее уверена.

```python
# Predictions in form of one-hot vectors (arrays of probabilities).
pd.DataFrame(predictions_one_hot)
```

```python
# Let's extract predictions with highest probabilites and detect what digits have been actually recognized.
predictions = np.argmax(predictions_one_hot, axis=1)
pd.DataFrame(predictions)
```

Итак, наша модель предсказывает, что первый пример из тестового набора равен 7.

```python
print(predictions[0])
```

Давайте напечатаем первое изображение из тестового набора, чтобы убедиться, что прогноз модели верен.

```python
plt.imshow(x_test_normalized[0], cmap=plt.cm.binary)
plt.show()
```

Мы видим, что наша модель сделала правильный прогноз и успешно распознала цифру «7». Давайте напечатаем еще несколько тестовых примеров и соответствующих прогнозов, чтобы увидеть, как модель работает и где она делает ошибки.

```python
numbers_to_display = 196
num_cells = math.ceil(math.sqrt(numbers_to_display))
fig, axes = plt.subplots(num_cells, num_cells, figsize=(15, 15))
axes = axes.flatten()  # превращаем двумерный массив осей в одномерный для удобства

for i in range(numbers_to_display):
    predicted_label = predictions[i]
    color_map = 'Greens' if predicted_label == y_test[i] else 'Reds'
    ax = axes[i]
    ax.imshow(x_test_normalized[i], cmap=color_map)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlabel(predicted_label)

plt.subplots_adjust(hspace=1, wspace=0.5)
plt.show()
```

## Построение матрицы коллизий (confusion matrix)

Вы можете видеть, что модель работает очень хорошо, но иногда (28 раз из 10000) она может спутать число «5» с «3» или число «2» с «3».

```python
confusion_matrix = tf.math.confusion_matrix(y_test, predictions)
f, ax = plt.subplots(figsize=(9, 7))
sn.heatmap(
    confusion_matrix,
    annot=True,
    linewidths=.5,
    fmt="d",
    square=True,
    ax=ax
)
plt.show()
```

### Загрузка данных и подготовка выборки


```python
mnist_dataset = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist_dataset.load_data()
```

### Оставляем для обучения первые 40,000 примеров (N = 40, N*1000)

```python
N = 40
x_train = x_train[:N*1000]
y_train = y_train[:N*1000]

print('x_train:', x_train.shape)
print('y_train:', y_train.shape)
print('x_test:', x_test.shape)
print('y_test:', y_test.shape)
```

### Исследование данных

```python
plt.imshow(x_train[2], cmap=plt.cm.binary)
plt.title(f'Цифра: {y_train[2]}')
plt.show()
```

### Выведем несколько примеров из обучающей выборки:

```python
numbers_to_display = 25
num_cells = math.ceil(math.sqrt(numbers_to_display))
plt.figure(figsize=(10, 10))
for i in range(numbers_to_display):
    plt.subplot(num_cells, num_cells, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i], cmap=plt.cm.binary)
    plt.xlabel(y_train[i])
plt.tight_layout()
plt.show()
```

### Нормализация данных

```python
x_train_normalized = x_train / 255.0
x_test_normalized = x_test / 255.0

# Для примера выводим нормализованное изображение:
plt.imshow(x_train_normalized[2], cmap=plt.cm.binary)
plt.title(f'Цифра: {y_train[2]} (нормализовано)')
plt.show()
```

### Построение базовой модели (MLP с 2 скрытыми слоями по 50 нейронов)


```python
model_baseline = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(
        units=50, 
        activation='relu',
        kernel_regularizer=tf.keras.regularizers.l2(0.002)
    ),
    tf.keras.layers.Dense(
        units=50, 
        activation='relu',
        kernel_regularizer=tf.keras.regularizers.l2(0.002)
    ),
    tf.keras.layers.Dense(10, activation='softmax')
])

model_baseline.summary()
```

# Для визуализации схемы модели:


```python
tf.keras.utils.plot_model(model_baseline, show_shapes=True, show_layer_names=True)
```

### Компиляция модели


```python
adam_optimizer = tf.keras.optimizers.legacy.Adam(learning_rate=0.001)
model_baseline.compile(
    optimizer=adam_optimizer,
    loss=tf.keras.losses.sparse_categorical_crossentropy,
    metrics=['accuracy']
)
```

### Обучение базовой модели

```python
log_dir = ".logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)

history_baseline = model_baseline.fit(
    x_train_normalized,
    y_train,
    epochs=10,
    validation_data=(x_test_normalized, y_test),
    callbacks=[tensorboard_callback]
)
```

### Визуализация результатов обучения базовой модели


```python
# График потерь
plt.figure(figsize=(8,4))
plt.plot(history_baseline.history['loss'], label='Training Loss')
plt.plot(history_baseline.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Базовая модель: Loss')
plt.legend()
plt.show()

# График точности
plt.figure(figsize=(8,4))
plt.plot(history_baseline.history['accuracy'], label='Training Accuracy')
plt.plot(history_baseline.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Базовая модель: Accuracy')
plt.legend()
plt.show()
```

### Оценка точности базовой модели

### Эксперимент: Улучшенная модель с Dropout


```python
model_improved = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(
        units=50,
        activation='relu',
        kernel_regularizer=tf.keras.regularizers.l2(0.002)
    ),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(
        units=50,
        activation='relu',
        kernel_regularizer=tf.keras.regularizers.l2(0.002)
    ),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(10, activation='softmax')
])

model_improved.summary()
```
